# K-Means Cluster

## Step 1- Importing Dependencies

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

## Step 2- loadind Dataset

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
x = data.data
```

## Step 3- Elbow Method

```python
from sklearn.cluster import KMeans
elbow = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    elbow.append(kmeans.inertia_)
```

### Visualize Elbow

```python
plt.plot(range(1,11),elbow)
plt.show()
```

## Step 4- Creating Model

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
y = kmeans.fit_predict(x)
```

### Visualization 

```python
plt.scatter(x[y == 0,0],x[y == 0,1],s=100,c='green',label='cluster 1')
plt.scatter(x[y == 1,0],x[y == 1,1],s=100,c='red',label='cluster 2')
plt.scatter(x[y == 2,0],x[y == 2,1],s=100,c='blue',label='cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='cyan',label='centroid')
plt.legend()
plt.show()
```
