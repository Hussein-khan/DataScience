# Seaborn

```python
# Importing libraries
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Built in dataset in seaborn library

```python
# load data set
tips = sns.load_dataset('tips')
tips.head()
```

## Visualize The Dataset

### RelPlot

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# set the theme
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',hue='smoker',style='smoker',size='size')
```

### Scatter Plot

```python
from sklearn.datasets import load_iris
dataset = load_iris()
data = pd.DataFrame(dataset.data,columns=dataset.feature_names)
data.head()
```

```python
data['target']=dataset.target
data.head()
```

```python
sns.scatterplot(data=data,x='sepal length (cm)',y='petal length (cm)',hue='target')
```

```python
sns.scatterplot(data=data,x='sepal width (cm)',y='petal width (cm)',hue='target')
```

## CountPlot

```python
data1 = sns.load_dataset('titanic')
data1
```

```python
sns.countplot(x='class',data=data1)
```

```python
sns.countplot(x='survived',data=data1)
```

## Bar Chart

```python
sns.barplot(x='sex',y='survived',hue='class',data=data1)
```

## Distribution Plot

```python
from sklearn.datasets import load_boston
bostan = load_boston()
ds = pd.DataFrame(bostan.data,columns=bostan.feature_names)
ds['price']=bostan.target
ds.head()
```

```python
sns.displot(ds['price'])
```

```python
sns.distplot(ds['price'])
```

```python
correlation = ds.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
