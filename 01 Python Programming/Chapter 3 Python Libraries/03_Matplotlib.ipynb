# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

Load some data 

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

## Plotting Data

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
# parabola with different color and symbol
plt.plot(x,y,'r*')
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

## Multiple Line Plot

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g+')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

## Bar Plot

```python
fig = plt.figure()
axis = fig.add_axes([0,0,1,1])
fruit = ['apple','bananas','orange','mangoes']
people = [80,70,75,95]
axis.bar(fruit,people)
plt.xlabel('fruits')
plt.ylabel('peoples')
plt.show()
```

## Pie Chart

```python
fig = plt.figure()
axis = fig.add_axes([0,0,1,1])
fruit = ['apple','bananas','orange','mangoes']
people = [80,70,75,95]
axis.pie(people,labels=fruit,autopct='%1.1f%%')
```

## Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
figur = plt.figure()
ax = figur.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

## 3D Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
```
